```python
from datasets import load_dataset
from googletrans import Translator
import pandas as pd

```

### Loading dataset from huggingface


```python
# Load the dataset
ds = load_dataset("ravikumar1478/Sentiment_analysis_on_Telugu_movie_Review_3")
df = ds['train'].to_pandas()
```

```python
translator = Translator()
retries = 3
```

### Translation of text using google translate api's

```python
def translate_text(text, src='te', dest='en'):
    for i in range(retries):
        try:
            translation = translator.translate(text, src=src, dest=dest)
            return translation.text
        except Exception as e:
            if i == retries - 1:
                return "Translation failed"
            continue

# Translate the reviews
df['translated_text'] = df['text'].apply(translate_text)
```

```python
translated_ds = {
    'text': df['translated_text'].tolist(),
    'label': df['label'].tolist()
}
```

```python
# Convert to DataFrame and save
translated_df = pd.DataFrame(translated_ds)
translated_df.to_csv('translated_telugu_reviews.csv', index=False)

# Preview results
print("\nSample Translations:")
for orig, trans in zip(df['text'][:3], translated_df['text'][:3]):
    print(f"\nOriginal (Telugu): {orig}")
    print(f"Translated (English): {trans}")
```
